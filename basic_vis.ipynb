Basic visualizations of MovieLens dataset using histograms. The following are visualized:
1. All ratings in the MovieLens Dataset.
2. All ratings of the ten most popular movies (movies which have received the most ratings).
3. All ratings of the ten best movies (movies with the highest average ratings).
4. All ratings of movies from three genres of your choice (create three separate visualizations).

```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
import itertools
```

```python
# load in data from pickle files

with open('get_movies.pkl', 'rb') as f:
    t = pickle.load(f)
    
movie_dic = t[0]
movie_genre_dic = t[1]

with open('get_ratings.pkl', 'rb') as f:
    lst = pickle.load(f)
```

```python
# Visualization of all ratings in the MovieLens Dataset
ratings = [t[2] for t in lst]

_ = plt.hist(ratings, bins=np.arange(0.5, 6.5, 1))
_ = plt.title('Histogram of All Ratings in MovieLens Dataset')
_ = plt.xlabel('Ratings')
_ = plt.ylabel('Frequency (Count)')
# plt.savefig('all_ratings_hist.png')
_ = plt.show()
```

```python
# Visualization of all ratings of ten most popular movies 
# (movies which have received the most ratings)

movie_rating_dic = {id: [] for id in movie_dic.keys()} # dic of movie id and ratings

for rating in lst:
    _, movie_id, r = rating
    movie_rating_dic[movie_id].append(r)
    
# dic of number of ratings for each movie
num_ratings = {movie: len(movie_rating_dic[movie]) for movie in movie_rating_dic}

# get top ten movies that received most number of ratings
top_10 = sorted(num_ratings.items(), key=lambda tup: tup[1], reverse=True)[:10]
top_10_ids = [t[0] for t in top_10]

# get list of all ratings of top ten movies to plot
top_10_ratings = []
for id in top_10_ids:
    top_10_ratings.append(movie_rating_dic[id])
top_10_ratings = list(itertools.chain.from_iterable(top_10_ratings))

_ = plt.hist(top_10_ratings, bins=np.arange(0.5, 6.5, 1))
_ = plt.title('Histogram of All Ratings of 10 Most Popular Movies')
_ = plt.xlabel('Ratings')
_ = plt.ylabel('Frequency (Count)')
# plt.savefig('most_pop_hist.png')
_ = plt.show()
```

```python
# Visualization of all ratings of the ten best movies 
# (movies with the highest average ratings)

# dic of average rating for each movie
avg_ratings = {movie: sum(movie_rating_dic[movie]) / len(movie_rating_dic[movie]) \
               for movie in movie_rating_dic}

# get ten best movies that received highest average rating
top_10 = sorted(avg_ratings.items(), key=lambda tup: tup[1], reverse=True)[:10]
top_10_ids = [t[0] for t in top_10]

# get list of all ratings of ten best movies to plot
top_10_ratings = []
for id in top_10_ids:
    top_10_ratings.append(movie_rating_dic[id])
top_10_ratings = list(itertools.chain.from_iterable(top_10_ratings))

_ = plt.hist(top_10_ratings, bins=np.arange(0.5, 6.5, 1))
_ = plt.title('Histogram of All Ratings of 10 Best Movies')
_ = plt.xlabel('Ratings')
_ = plt.ylabel('Frequency (Count)')
# plt.savefig('best_hist.png')
_ = plt.show()
```

```python
# Visualization of all ratings of movies from three genres of 
# your choice (create three separate visualizations).

# Fantasy
fant_movies = movie_genre_dic['Fantasy']

# get list of all ratings of action movies to plot
fant_ratings = []
for id in fant_movies:
    fant_ratings.append(movie_rating_dic[id])
fant_ratings = list(itertools.chain.from_iterable(fant_ratings))

_ = plt.hist(fant_ratings, bins=np.arange(0.5, 6.5, 1))
_ = plt.title('Histogram of All Ratings of Fantasy Movies')
_ = plt.xlabel('Ratings')
_ = plt.ylabel('Frequency (Count)')
# plt.savefig('action_hist.png')
_ = plt.show()
```
